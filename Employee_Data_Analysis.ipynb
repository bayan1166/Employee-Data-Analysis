```python
text_columns = ['Job Title', 'Department', 'Business Unit', 'Gender', 'Ethnicity', 'Country', 'City']

for col in text_columns:
    if df[col].dtype == 'object':
        df[col] = df[col].str.strip().str.title()
```

هون بكل التكست كولمنز انا شلت السبيسز

```python
df['EEID'] = df['EEID'].astype(str).str.strip()

counts = df.groupby('EEID').cumcount()

df['EEID'] = df['EEID'].where(
    counts == 0,
    df['EEID'] + "_" + (counts + 1).astype(str)
)

remaining_duplicates = df.duplicated(subset=['EEID']).sum()
print(f"Remaining duplicate EEIDs: {remaining_duplicates}")
```

## TASK5

```python
department_summary = df.groupby('Department').agg(
    Average_Age=pd.NamedAgg(column='Age', aggfunc='mean'),
    Average_Salary=pd.NamedAgg(column='Annual Salary', aggfunc='mean'),
    Headcount=pd.NamedAgg(column='EEID', aggfunc='count')
).reset_index()
```

```python
department_summary = department_summary.sort_values(by='Headcount', ascending=False)
```

```python
print("Department Summary (Average Age, Average Salary, and Headcount):")
display(department_summary)
```

## TASK6

```python
dept_eth_summary = df.groupby(['Department', 'Ethnicity']).agg(
    Max_Age=pd.NamedAgg(column='Age', aggfunc='max'),
    Min_Age=pd.NamedAgg(column='Age', aggfunc='min'),
    Median_Salary=pd.NamedAgg(column='Annual Salary', aggfunc='median')
).reset_index()

print("1. Max Age, Min Age, and Median Salary per Dept & Ethnicity:")
display(dept_eth_summary.head())
```

```python
salary_pivot = pd.pivot_table(
    df, 
    values='Annual Salary', 
    index='Department', 
    columns='Gender', 
    aggfunc='mean'
)
```

```python
if 'Male' in salary_pivot.columns and 'Female' in salary_pivot.columns:
    salary_pivot['Pay Gap (%)'] = ((salary_pivot['Male'] - salary_pivot['Female']) / salary_pivot['Male']) * 100

print("\n2. Pivot Table (Average Salary by Dept x Gender) + Pay Gap:")
display(salary_pivot)
```

```python
headcount_crosstab = pd.crosstab(df['Department'], df['Ethnicity'])

print("\n3. Crosstab (Headcount by Department x Ethnicity):")
display(headcount_crosstab)
```

```python
df['Dept_Avg_Salary'] = df.groupby('Department')['Annual Salary'].transform('mean')
df['Salary Deviation'] = df['Annual Salary'] - df['Dept_Avg_Salary']
```

```python
print("\n4. First 5 rows showing Salary Deviation using .transform():")
display(df[['EEID', 'Department', 'Annual Salary', 'Dept_Avg_Salary', 'Salary Deviation']].head())
```

## TASK7

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
df['Status'] = np.where(df['Exit Date'].isna(), 'Active', 'Terminated')
```

```python
today = pd.Timestamp.today()
end_date_for_calc = df['Exit Date'].fillna(today)

df['Tenure_Years'] = ((end_date_for_calc - df['Hire Date']).dt.days / 365.25).round(2)
```

```python
df['Hire_Year'] = df['Hire Date'].dt.year
df['Hire_Quarter'] = df['Hire Date'].dt.quarter
df['Hire_DayOfWeek'] = df['Hire Date'].dt.day_name()

print("1, 2, 3. Datetime features created successfully:")
display(df[['EEID', 'Hire Date', 'Exit Date', 'Status', 'Tenure_Years', 'Hire_Year', 'Hire_Quarter', 'Hire_DayOfWeek']].head())
```

```python
df['Exit_Year'] = df['Exit Date'].dt.year
hires_per_year = df['Hire_Year'].value_counts().sort_index()
exits_per_year = df['Exit_Year'].value_counts().sort_index()
trend_df = pd.DataFrame({'Hires': hires_per_year, 'Exits': exits_per_year}).fillna(0)
```

```python
plt.figure(figsize=(10, 5))
plt.plot(trend_df.index, trend_df['Hires'], marker='o', label='Hires', color='green', linewidth=2)
plt.plot(trend_df.index, trend_df['Exits'], marker='x', label='Exits', color='red', linewidth=2)
plt.title('Hiring and Attrition Trends Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Employees', fontsize=12)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import os
```

```python
output_dir = "Charts_Output"
os.makedirs(output_dir, exist_ok=True)
```

```python
# 1. Bar Chart: Average Salary by Department
plt.figure(figsize=(10, 6))
dept_salary = df.groupby('Department')['Annual Salary'].mean().sort_values(ascending=False)
plt.bar(dept_salary.index, dept_salary.values, color='skyblue', edgecolor='black')
plt.title('Average Annual Salary by Department (Matplotlib)')
plt.xlabel('Department')
plt.ylabel('Average Salary ($)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'{output_dir}/1_Avg_Salary_Bar.png')
plt.show()
```

```python
# 2. Histogram: Age Distribution
plt.figure(figsize=(8, 5))
plt.hist(df['Age'].dropna(), bins=15, color='lightgreen', edgecolor='black')
plt.title('Age Distribution of Employees (Matplotlib)')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.75)
plt.tight_layout()
plt.savefig(f'{output_dir}/2_Age_Distribution_Hist.png')
plt.show()
```

```python
# 3. Line Chart: Hires per Year
plt.figure(figsize=(10, 5))
hires_per_year = df['Hire_Year'].value_counts().sort_index()
plt.plot(hires_per_year.index, hires_per_year.values, marker='o', linestyle='-', color='purple', linewidth=2)
plt.title('Hiring Trends Over Time (Matplotlib)')
plt.xlabel('Year')
plt.ylabel('Number of Hires')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig(f'{output_dir}/3_Hiring_Trends_Line.png')
plt.show()
```

```python
sns.set_theme(style="whitegrid")
```

```python
# 4. Boxplot: Salary by Department (Spot Spread/Outliers)
plt.figure(figsize=(12, 6))
sns.boxplot(x='Department', y='Annual Salary', data=df, palette='Set2')
plt.title('Salary Distribution & Outliers by Department (Seaborn)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'{output_dir}/4_Salary_Outliers_Boxplot.png')
plt.show()
```

```python

# 5. Correlation Heatmap: Age, Salary, Bonus %, Tenure
plt.figure(figsize=(8, 6))
corr_cols = ['Age', 'Annual Salary', 'Bonus %', 'Tenure_Years']
corr_matrix = df[corr_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1, linewidths=0.5)
plt.title('Correlation Heatmap (Seaborn)')
plt.tight_layout()
plt.savefig(f'{output_dir}/5_Correlation_Heatmap.png')
plt.show()
```

```python

# 6. Grouped Bar Chart: Average Salary by Department and Gender
plt.figure(figsize=(12, 6))
sns.barplot(x='Department', y='Annual Salary', hue='Gender', data=df, palette='muted', errorbar=None)
plt.title('Average Salary by Department and Gender (Seaborn)')
plt.legend(title='Gender')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig(f'{output_dir}/6_Grouped_Bar_Salary_Gender.png')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Annual Salary', hue='City', size='Tenure_Years', sizes=(20, 200), data=df, alpha=0.7)
plt.title('Bonus: Age vs Salary by City (Sized by Tenure)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.savefig(f'{output_dir}/7_Bonus_Scatter_Plot.png')
plt.show()
```
